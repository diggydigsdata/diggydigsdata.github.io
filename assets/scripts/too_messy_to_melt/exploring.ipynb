```python
import pandas as pd
```

```python
gold_data = pd.read_parquet("../../files/datasets/too_messy_to_melt/gold_relegation_attendance.parquet")
print(gold_data.head())
```

```python
gold_data["Relegation_Season_Index"] = (
    gold_data["Relegation_Event_ID"].astype(str)
    + " | Y"
    + gold_data["Year_vs_Relegation"].astype(str)
)
```

```python
assert gold_data["Relegation_Season_Index"].is_unique
```

```python
dupes = (
    gold_data
    .groupby(["Relegation_Event_ID", "Year_vs_Relegation"])
    .size()
    .reset_index(name="n")
    .query("n > 1")
    .sort_values("n", ascending=False)
)

print(dupes.head(100))
```

```python
silver_data = pd.read_parquet("../../files/datasets/too_messy_to_melt/silver_relegation_attendance.parquet")
print(silver_data.head())
dupes_silver = (
    silver_data
    .groupby(["Relegation_Event_ID", "Year_vs_Relegation"])
    .size()
    .reset_index(name="n")
    .query("n > 1")
    .sort_values("n", ascending=False)
)

print(dupes_silver.head(100))
```
